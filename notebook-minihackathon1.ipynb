```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
```

```python
imbd_dataset = pd.read_csv('/kaggle/input/imdb-dataset-of-top-1000-movies-and-tv-shows/imdb_top_1000.csv')
conn = sqlite3.connect('/kaggle/working/imbd_dataset.db')
imbd_dataset.to_sql('imbd_dataset', conn, if_exists='replace', index=False)
```

```python
%sql sqlite:///imbd_dataset.db
```

****Analyze how movies directed by different directors perform in terms of gross earnings. Are there any noticeable trends or patterns?

```python
%%sql
SELECT Director AS director_name, Genre AS genre, SUM(Gross) AS total_gross
FROM imbd_dataset
GROUP BY Director, Genre
ORDER BY total_gross DESC;
```

**Investigate how different genres have evolved in popularity over time.**

```python
%%sql
SELECT 
    Released_Year AS Year, 
    Genre, 
    COUNT(*) AS Count
FROM imbd_dataset
GROUP BY Released_Year, Genre
ORDER BY Year, Count ASC;
```

**Explore if there's a relationship between a movie's IMDB rating and its box office earnings.**

```python
%%sql
SELECT 
    IMDB_Rating, 
    Gross
FROM imbd_dataset    
WHERE 
    IMDB_Rating IS NOT NULL 
    AND Gross IS NOT NULL
ORDER BY 
    IMDB_Rating ASC;
```

** Examine whether the duration of a movie influences its ratings or financial success.**

```python
%%sql
SELECT 
    Runtime AS movie_duration, 
    Gross,
    IMDB_Rating
FROM  imbd_dataset  
    
WHERE 
    Runtime IS NOT NULL 
    AND IMDB_Rating IS NOT NULL
    AND  Gross IS NOT NULL
ORDER BY 
    movie_duration ASC;
```
